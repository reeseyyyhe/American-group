### In this file, we can reorder the matching records from datasets to help us more easily verify the matching correctness

```python
import pandas as pd
```

```python
ds1_with_matchings = pd.read_csv('C:/Columbia_University/Research/History_African_American_Migration_Pattern/US_MLP_datasets/1850_matching.csv')
ds2_with_matchings = pd.read_csv('C:/Columbia_University/Research/History_African_American_Migration_Pattern/US_MLP_datasets/1860_matching.csv')
```

```python
ds1_with_matchings_ordered = ds1_with_matchings.sort_values(by = "matching_id")
ds1_with_matchings_ordered
```

```python
ds2_with_matchings_ordered = ds2_with_matchings.sort_values(by = "matching_id")
ds2_with_matchings_ordered
```

```python
ds1_with_matchings_ordered.to_csv('C:/Columbia_University/Research/History_African_American_Migration_Pattern/US_MLP_datasets/1850_matching.csv', index = False)
```

```python
ds2_with_matchings_ordered.to_csv('C:/Columbia_University/Research/History_African_American_Migration_Pattern/US_MLP_datasets/1860_matching.csv', index = False)
```
