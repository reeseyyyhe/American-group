```python
temp_df = df[(df['Census.Year']=='Ontario, Canada, Deaths and Deaths Overseas 1869-1947') \
                         & (df['CalculatedBirthYear']!='F')]
for i in temp_df.index:
    df.loc[i, 'ID'] = 'Matson: Ontario, Canada, Deaths and Deaths Overseas 1869-1947'
    df.loc[i, 'Census.Year'] = df.loc[i, 'CalculatedBirthYear'] + df.loc[i, 'Age']
df[(df['ID']=='Matson: Ontario, Canada, Deaths and Deaths Overseas 1869-1947')].head(3)
```

```python
# appear once list
census_year_list = [
    'Virginia, Death Records 1912-2014',
    'Ontario, Canada Voter Lists 1867-1900',
    '1855 Massachussetts State Census',
    'Canada, City and Area Directories, 1819-1906, 1867 Directory',
    'U.S. Civil War Records and Soldier Profiles 1861-1865',
    'Find-A-Grave Index', 
    'New York State Census 1892', 
    'Canada, City and Area Directories, 1819-1906',
    'Cook County, Illinois, Deaths Index, 1878-1932',
    'Michigan Marriage Records 1867-1952' ,
    'Ontario, Canada, Deaths and Deaths Overseas, 1869-1947' 
]
```

```python
for year in census_year_list:
    idx = df.loc[df['Census.Year']==year].index.tolist()[0]
    birth_year = df.loc[idx, 'CalculatedBirthYear']
    age = df.loc[idx, 'Age']
    ID = df.loc[idx, 'ID']
    # update
    #print(ID+": "+year)
    df.loc[idx, 'ID'] = str(ID+": "+year)
    if birth_year not in ['-', 'nan'] and age not in ['-', 'nan']:
        #print(birth_year+age)
        df.loc[idx, 'Census.Year'] = birth_year + age
    else: # canot calcualte the census year
        print(year)
```
